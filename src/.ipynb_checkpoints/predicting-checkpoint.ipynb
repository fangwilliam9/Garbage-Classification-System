```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import matplotlib.image as mpimg
from collections import Counter

def load_model(model_path):
    """加载训练好的模型"""
    if not os.path.exists(model_path):
        raise FileNotFoundError(f"Model file not found: {model_path}")
    return keras.models.load_model(model_path)

def get_image_files(directory):
    """获取指定目录下的所有图片文件"""
    image_files = [f for f in os.listdir(directory) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]
    if not image_files:
        raise FileNotFoundError(f"No image files found in the directory: {directory}")
    return image_files

def get_user_input(image_files):
    """获取用户输入的图片索引"""
    while True:
        try:
            user_input = input(f'Enter the index of the image to test (0 to {len(image_files) - 1}): ').strip()
            if not user_input:
                print('Input is empty. Please enter a valid integer.')
                continue

            ind = int(user_input)
            if ind < 0 or ind >= len(image_files):
                print(f'Invalid index. Please enter a value between 0 and {len(image_files) - 1}.')
                continue

            return ind

        except ValueError:
            print('Invalid input. Please enter a valid integer.')

def load_and_preprocess_image(image_path):
    """加载并预处理图片"""
    img = mpimg.imread(image_path)
    if img is None:
        raise ValueError(f"Failed to load image: {image_path}")

    # 将图片转换为可写的 NumPy 数组
    img = np.copy(img)  # 创建可写副本

    # 将图片转换为模型输入格式
    x = keras.preprocessing.image.img_to_array(img)
    x = np.copy(x)  # 再次确保数组可写

    # 调整图片大小并确保数组可写
    x = tf.image.resize(x, (224, 224)).numpy()  # 转换为 NumPy 数组
    x = np.copy(x)  # 再次确保数组可写

    # 扩展维度并预处理
    x = np.expand_dims(x, axis=0)
    x = keras.applications.mobilenet_v2.preprocess_input(x)  # 使用与训练相同的预处理
    return x, img

def load_class_names(dataset_dir):
    """从数据集目录加载有效的类别名称"""
    class_names = [f for f in os.listdir(dataset_dir) 
                   if os.path.isdir(os.path.join(dataset_dir, f)) and not f.startswith('.')]
    if not class_names:
        raise ValueError(f"No valid class directories found in: {dataset_dir}")
    return sorted(class_names)

def predict_and_display(model, image, img, class_names):
    """使用模型进行预测并显示结果"""
    predictions = model.predict(image)
    print("Prediction probabilities:", predictions)  # 打印预测概率分布

    pred_class = class_names[np.argmax(predictions)]  # 使用动态加载的类别名称

    # 显示结果
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.title(f'Predicted: {pred_class}')
    plt.show()

def main():
    try:
        # 加载模型
        model = load_model('classification_model.keras')

        # 获取当前工作目录
        current_dir = os.getcwd()

        # 获取当前目录下的图片文件
        image_files = get_image_files(current_dir)

        # 显示当前目录下的图片列表
        print('Available images in the current directory:')
        for i, image_file in enumerate(image_files):
            print(f'{i}: {image_file}')

        # 获取用户输入的图片索引
        ind = get_user_input(image_files)

        # 构建图片路径
        image_path = os.path.join(current_dir, image_files[ind])

        # 加载并预处理图片
        image, img = load_and_preprocess_image(image_path)

        # 动态加载类别名称
        # 如果类别名称未保存到模型中，从数据集的目录结构中加载
        dataset_dir = '../dataset/Garbage classification/Garbage classification'  # 替换为你的数据集目录
        class_names = load_class_names(dataset_dir)
        print(f"Loaded class names from dataset directory: {class_names}")

        # 使用模型进行预测并显示结果
        predict_and_display(model, image, img, class_names)

    except Exception as e:
        print(f'An error occurred: {e}')

if __name__ == '__main__':
    main()
```
